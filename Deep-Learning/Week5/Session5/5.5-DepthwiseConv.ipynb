# MobileNet Depthwise Concatenation

```python
from keras.models import Sequential
from keras.layers import DepthwiseConv2D, Conv2D
from tensorflow.keras.utils import plot_model
```

```python
# Define input dimensions
height = 224
width = 224
channels = 3  # Assuming RGB images
num_filters = 64  # Number of filters for the pointwise convolution

# Define a Sequential model
model = Sequential()

# Add a DepthwiseConv2D layer
model.add(DepthwiseConv2D(kernel_size=(3, 3), strides=(1, 1), padding='same', activation='relu', input_shape=(height, width, channels)))

# Add a pointwise convolution
model.add(Conv2D(filters=num_filters, kernel_size=(1, 1), strides=(1, 1), padding='same', activation='relu'))

# Add more layers as needed

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Print model summary
model.summary()
```

```python
plot_model(model,show_shapes = True, show_layer_names = True)
```
