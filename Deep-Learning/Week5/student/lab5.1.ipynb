# Lab 5.1 Morphological Operations with Skimage

```python
from copy import deepcopy

import matplotlib.pyplot as plt
import numpy as np
from skimage import morphology, data, img_as_ubyte
```

## Demo

### Dilation Exercise

```python
# Load image
image = data.coins()
```

```python
# Define structuring element
selem = morphology.square(10)
```

```python
# Apply dilation operation to image
dilated_image = morphology.dilation(image, selem)
```

```python
# Display original and dilated images
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))
ax = axes.ravel()
ax[0].imshow(image, cmap='gray')
ax[0].set_title('Original')
ax[1].imshow(dilated_image, cmap='gray')
ax[1].set_title('Dilated')
plt.show()
```

### Erosion Exercise

```python
# Load image
image = data.coins()
```

```python
# Define structuring element
selem = morphology.square(5)
```

```python
# Apply erosion operation to image
eroded_image = morphology.erosion(image, selem)
```

```python
# Display original and eroded images
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))
ax = axes.ravel()
ax[0].imshow(image, cmap='gray')
ax[0].set_title('Original')
ax[1].imshow(eroded_image, cmap='gray')
ax[1].set_title('Eroded')
plt.show()
```

### Opening Exercise

```python
# Load image
image = data.coins()

noisy_image = img_as_ubyte(data.coins())
noise = np.random.random(noisy_image.shape)
noisy_image[noise > 0.99] = 255
noisy_image[noise < 0.01] = 0
```

```python
# Define structuring element
selem = morphology.square(1)
```

```python
# Apply opening operation to image
opened_image = morphology.opening(image, selem)
```

```python
# Display original and opened images
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))
ax = axes.ravel()
ax[0].imshow(image, cmap='gray')
ax[0].set_title('Original')
ax[1].imshow(noisy_image, cmap='gray')
ax[1].set_title('Noisy')
ax[2].imshow(opened_image, cmap='gray')
ax[2].set_title('Opened')
plt.show()
```

### Closing Exercise

```python
# Load image
image = data.coins()

noisy_image = deepcopy(image)
noisy_image[10:15, 30:35] = 1
```

```python
# Define structuring element
selem = morphology.square(10)
```

```python
# Apply closing operation to image
closed_image = morphology.closing(noisy_image, selem)
```

```python
# Display original and closed images
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))
ax = axes.ravel()
ax[0].imshow(image, cmap='gray')
ax[0].set_title('Original')
ax[1].imshow(noisy_image, cmap='gray')
ax[1].set_title('Noisy')
ax[2].imshow(closed_image, cmap='gray')
ax[2].set_title('Closed')
plt.show()
```

#### What is a structuring element selem? What is its purpose? How to choose the size? 
A structuring element selem is a small binary image that is used as a probe to analyze the geometrical properties of an image. It is typically a small square or rectangle but can be any shape. The size of the structuring element is determined by the size of the objects in the image that you want to analyze. A larger structuring element will analyze larger objects, while a smaller structuring element will analyze smaller objects. The choice of the size of the structuring element depends on the specific task and the characteristics of the image being analyzed.
