1. Create an AudioFile class with the following:
    - Passed in attributes: title, artist, duration_in_seconds
    - Default attributes: play_count = 0, like = False
    - Write a constructor to initialize the passed in and default attributes.
2. Define the following methods:
    - print_audio_details that prints the title, artist and play_count in a readerfriendly way like "Imagine by The Beatles | Play Count: 0"
    - heart that will change the like value to True
    - unheart that will change the like value to False
    - play that will increase the play_count by 1

```python
class AudioFile(object):
    """
    Class representing a Audio file representing the 
    """
    def __init__(self, title, artist, duration_in_seconds):
        """
        params:
            title: title of audio file
            artist: artist related to audio file
            duration_in_seconds: duration of audio file in seconds
        """
        self.title = title
        self.artist = artist 
        assert isinstance(duration_in_seconds, int), f"duration_in_seconds should be integer"
        self.duration_in_seconds = duration_in_seconds
        self.play_count = 0 
        self.like = False
    
    def print_audio_details(self):
        """Method prints the details of the audio like title, artist and play count"""
        print(f"{self.title} by The {self.artist} | Play Count : {self.play_count}")
    
    def heart(self):
        """Method changes the boolean variable like True"""
        self.like = True
    
    def unheart(self):
        """Method changes the boolean variable like to False"""
        self.like = False
    
    def play(self):
        """Method increases the play count by 1"""
        self.play_count+=1
```

```python
## Running Examples to demonstrates class attributes and methods

# Creating an instance of AudioFile Class
audio1 = AudioFile(title = "Imagine", artist = 'Beatles', duration_in_seconds = 120)
# Calling class method print_audio_details
audio1.print_audio_details()
print()
# Using class attribute "like"
print(f"Marked Heart : {audio1.like}\n")

# Demonstrating the use of "play" method
for cnt in range(3):
    audio1.play()
    audio1.print_audio_details()
print()

audio1.heart()
print(f"Marked Heart : {audio1.like}\n")

audio1.unheart()
print(f"Marked Heart : {audio1.like}\n")
```

3. Create a subclass called Song that inherits from the AudioFile class, and has:

```python
class Song(AudioFile):
    """
    Child Class of AudioFile
    Inherit all the methods of AudioFile
    One addional method added : words_per_minute
    """
    def __init__(self, title, artist, duration_in_seconds, lyrics):
        """
        params:
            title: title of Song
            artist: artist related to Song
            duration_in_seconds: duration of Song in seconds
            lyrics: lyrics of the Song
        """
        # Initialising the base attributes using the parent class
        super().__init__(title, artist, duration_in_seconds)
        self.lyrics = lyrics
    
    def words_per_minute(self):
        """The method calculates words per minute in the song, using total duration and lyrics"""
        number_of_words = len(self.lyrics.split( ))
        total_minutes = self.duration_in_seconds / 60
        wpm = int(number_of_words // total_minutes)
        return wpm
```

4. Create a subclass called Podcast that inherits from the AudioFile class, and has:

```python
class Podcast(AudioFile):
    """
    Child Class of AudioFile
    Inherit all the methods of AudioFile
    One addional method added : words_per_minute
    """
    def __init__(self, title, artist, duration_in_seconds, transcript):
        """
        params:
            title: title of Song
            artist: artist related to Song
            duration_in_seconds: duration of Song in seconds
            transcript: Transcript of the Podcast
        """
        super().__init__(title, artist, duration_in_seconds)
        self.transcript = transcript
    
    def words_per_minute(self):
        """The method calculates words per minute in the Podcast, using total duration and lyrics"""
        number_of_words = len(self.transcript.split( ))
        total_minutes = self.duration_in_seconds / 60
        wpm = int(number_of_words // total_minutes)
        return wpm
```

8. Create a Playlist class that has a playlist_name attribute and takes in a list of AudioFiles
    - Create a constructor
    - Create a method called print_playlist_details that prints the playlist_name and lists the audio files it contains in a reader-friendly way
    
Create a new playlist instance, add a few songs and print the details

```python
class Playlist(object):
    """
    A class demonstrating the concept of playlist 
    Has two class attributes 
        a) playlist name : name of the playlist
        b) audio_files_list : list of audio files in playlist'
    Has one class method
        a) print_playlist_details : print details of the playlist
    """
    def __init__(self, playlist_name, audio_files_list = []):
        self.playlist_name = playlist_name
        self.audio_files_list = audio_files_list
    
    def print_playlist_details(self):
        """Loop over all audio files and call their print_audio_details method to print details"""
        print(f"Playlist Name : {self.playlist_name}\n")
        if len(self.audio_files_list) == 0:
            print(f"No audio Files added currently to the playlist")
        else:
            print(f"{len(self.audio_files_list)} Audio Files Found. List of added Audio Files\n")
            for idx, audio in enumerate(self.audio_files_list):
                audio.print_audio_details()    
```

```python
aud1 = AudioFile(title = "Demons", artist = 'Imagine Dragons', duration_in_seconds = 150)
aud2 = AudioFile(title = "Loose Yourself", artist = 'Eminem', duration_in_seconds = 240)
aud3 = AudioFile(title = "Numb", artist='Lincoln Park', duration_in_seconds = 130)

audio_files = [aud1, aud2, aud3]

playlist1 = Playlist(playlist_name = 'Playlist1', audio_files_list = audio_files)
playlist1.print_playlist_details()
```

### 2.What was the total spent on salaries by each team, each year? What were the top 3 highest spend teams / years and the bottom 3 highest spend teams / years?

```python
# Calculating salaries per year per team by grouping the df_salaries over yearID and teamID
salaries_per_year_per_team = df_salaries.groupby(['yearID', 'teamID'])['salary'].sum()
salaries_per_year_per_team
```

```python
# Formatting the above result into a dataframe format
year_id_list = []
team_id_list = []
salary_list = []
for key in salaries_per_year_per_team.keys():
    year_id_list.append(key[0])
    team_id_list.append(key[1])
    salary_list.append(salaries_per_year_per_team[key])
df_salaries_per_year_per_team = pd.DataFrame(columns=['yearID', 'teamID', 'salary'])
df_salaries_per_year_per_team.yearID = year_id_list
df_salaries_per_year_per_team.teamID = team_id_list
df_salaries_per_year_per_team.salary = salary_list
df_salaries_per_year_per_team.head()
```

```python
# Getting top 3 highest spenders by sorting the above dataframr on column salary in desc order
# Top top 3 for highest spenders
top_3_highest_spenders = df_salaries_per_year_per_team.sort_values(by=['salary'], ascending=False)[:3]
top_3_highest_spenders 
```

```python
# Getting top 3 highest spenders by sorting the above dataframr on column salary in asc order
# Top top 3 for lowest spenders
bottom_3_highest_spenders = df_salaries_per_year_per_team.sort_values(by=['salary'], ascending=True)[:3]
bottom_3_highest_spenders
```

### 4. For each team, what percent of players bat right vs left vs both?

```python
# Merging team df_master and df_salaries using inner join on playerID
# This kind of merging will help to map players to different teams during diff years
df_team_player = pd.merge(df_master, df_salaries, how='inner', left_on=['playerID'], right_on=['playerID'])
df_team_player.head()
```

```python
# Calcutating Total count of players after grouping the df_team_player on teamID and bats
bat_player_count_per_team = df_team_player.groupby(['teamID', 'bats'])['playerID'].count()

team_dict = {}

# Calcuating percentage of Left, Right and Both handed Bats 
for key in bat_player_count_per_team.keys():
    teamID, bat = key
    count = bat_player_count_per_team[key]
    if teamID not in team_dict:
        team_dict[teamID] = {}
    team_dict[teamID][bat] = count 
    
    if 'R' in team_dict[teamID] and 'L' in team_dict[teamID] and 'B' in team_dict[teamID]:
        team_dict[teamID]['total'] = team_dict[teamID]['R'] + team_dict[teamID]['L'] + team_dict[teamID]['B']
        team_dict[teamID]['L_Per'] = round(team_dict[teamID]['L'] / team_dict[teamID]['total'] * 100,2)
        team_dict[teamID]['R_Per'] = round(team_dict[teamID]['R'] / team_dict[teamID]['total'] * 100,2)
        team_dict[teamID]['B_Per'] = round(team_dict[teamID]['B'] / team_dict[teamID]['total'] * 100,2)

df_bat_per = pd.DataFrame(team_dict).T
df_bat_per.head(10)
```

### 5. Which 10 players have the longest careers? Assume that the debut and finalGame columns comprise the start and end, respectively, of a player's career.

```python
# Checking for null debut dates and finalGames
num_null_debuts = sum(df_master.debut.isnull())
num_null_finalGame = sum(df_master.finalGame.isnull())
print(f"Number of Null Debut Dates:{num_null_debuts}")
print(f"Number of Null FinalGame Dates:{num_null_finalGame}")

# Dropping the null debut rows and creating new dataframe 
df_master_fix = df_master[~df_master.debut.isnull()]
print(f"Orginal df_master Size: {df_master.shape[0]}")
print(f"Fixed df_master Size: {df_master_fix.shape[0]}")
```

```python
def fix_dates(date):
    """Utility function to fix the discrepancy in date formatting"""
    if len(date.split('-')) == 3:
        date = date.replace('-','/')
    
    if len(date.split('/')[0]) == 4:
        year, month, date = date.split('/')
        date = f"{month}/{date}/{year}"
    
    return date

# Fixing date discrepancies for debut and finalGame
df_master_fix['debut'] = df_master_fix['debut'].apply(lambda x : fix_dates(x))
df_master_fix['finalGame'] = df_master_fix['finalGame'].apply(lambda x : fix_dates(x))
# Finding careerLen by subtracing finalGame and debut
df_master_fix['careerLen'] = (pd.to_datetime(df_master_fix['finalGame']) - pd.to_datetime(df_master_fix['debut']))
df_master_fix['careerLen'] = df_master_fix['careerLen'].apply(lambda x : x.days)
df_master_fix.head()
```

```python
# Finding top 10 longest career players by sorting the dataframe on careerLen
top_10_longest_career_players = df_master_fix.sort_values(by=['careerLen'], ascending=False)[:10]
top_10_longest_career_players[['playerID', 'nameGiven', 'careerLen']]
```
